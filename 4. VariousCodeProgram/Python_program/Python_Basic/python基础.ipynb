```python
# 非对称布局
from matplotlib import gridspec
import matplotlib.pyplot as plt 
# 准备数据
t = np.arange(1, 100, 2)
fig = plt.figure(figsize=(12,12))    
gs = gridspec.GridSpec(3,3)           #  将画布分成三行, 三列
gs.update(wspace=0.2, hspace=0.3)     # 设置水平间隔 和 垂直间隔
ax = {}
ax['t'] = fig.add_subplot(gs[0,:2])   # 设置第一个图的位置是  第一行, 并且占据了两列 
ax['t'].plot(10 +10*np.cos(t) , color= 'r' , label = 'World')
ax['t'].legend(ncol= 1 )              # 将图例设置为 1 列

ax['pr'] = fig.add_subplot(gs[1, :2], sharex=ax['t'] )
ax['pr'].set_title('Precipitation', loc='left', fontsize=14)

ax['hist_t']= fig.add_subplot(gs[0,2], sharey=ax['t'])
ax['hist_pr'] = fig.add_subplot(gs[1,2], sharey = ax['pr'], sharex= ax['pr'])

ax['hist_t'].hist(t, 
                  orientation='horizontal',
                  density=True, 
                  bins=20,
                  label = "Hello",  
                 )
ax['hist_t'].legend(frameon= False, ncol=1,   loc = 'upper right',  )

ax['hist_t'].tick_params(axis = 'y', labelleft=False)   # tick_params参数的作用: 调整坐标轴的刻度,  此处不显示y轴的刻度. 
```

```python
# 自定义绘图
def plot_geo_mean(da,lat_min, lat_max, lon_min , lon_max, 
                     lat_name='lat', lon_name='lon', time_name = 'time', 
                     var_name = 'SST', var_unit=r"$^{\circ}$C",
                     title= None, 
                     figsize=(12, 8), clon = 180) :
    # calculation 
    mask_lat = (da[lat_name] >=latmin) & (da[lat_name]<=lat_max)
    mask_lon = (da[lon_name] >=lon_min) & (da[lon_name]<=lon_max)
    dac  = da.sel(
        { 
            lat_name : da[lat_name][mask_lat], 
            lon_name : da[lon_name][mask_lon], 
        }
    )
    wgts = np.cos(np.deg2rad(dac[lat_name]))
    m = dac.weighted(wgts).mean((lon_name, lat_name))

    # visulation 
    # map 
    fig = plt.figure(figsize = figsize)
    gs = gridspec.GridSpec(2,3)
    gs.update(wspace = 0.1, hspace = 0.1)
    ax = {}
    ax['map'] = fig.add_subplot(gs[:1,:2],
                                projection = ccrs.Robinson(central_longitude = clon))
    ax['map'].set_global()
    ax['map'].stock_img()
    ax['map'].coastlines()
    if title is not None : 
        ax['map'].set_title(title)

    # polygon
    p1_lon , p1_lat = lon_min, lat_min
    p2_lon, p2_lat = lon_max, lat_min 
    p3_lon, p3_lat = lon_max, lat_max
    p4_lon, p4_lat = lon_min , lat_max 
    lons = [p1_lon, p2_lon, p3_lon, p4_lon]
    lats = [p1_lat, p2_lat, p3_lat, p4_lat]
    ax['map'].plot(
                    lons, lats, marker = 'o' , color = 'coral', 
                    transform= ccrs.PlateCaree()
    )
    ax['map'].fill(lons, lats, color='coral',
                   tranform= ccrs.geodetic(),
                   alpha=0.4
    )

    # 后面就不写了
```
